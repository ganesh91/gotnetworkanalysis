# Data Processing - Nodes

## Taking the data from relations.csv into pandas to find unique charaters to build nodes!

```python
import numpy as np
import pandas as pd
```

```python
relations_df = pd.read_csv('relations.csv', header=None, quotechar="\"", encoding="latin_1")
```

```python
relations_df.head()
```

## Making a list of unique characters of 'Game of Thrones'

```python
characters=list(relations_df[2].unique())+(list(relations_df[0].unique()))
```

```python
print(characters)
```

## Creating nodes.csv which contains data in required format to communication with Frontend Javascript

```python
nodes=open('nodes.csv','w',encoding="latin_1")
for char in characters:
    nodes.write("{data:{id: '" + str(char).replace("'","\\\'").replace(" ","") + "', lbl :'" + str(char).replace("'","\\\'") + "'}}," + "\n")
nodes.close()
```
