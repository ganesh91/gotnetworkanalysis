# Data Cleaning - Relationship: Family

## Extracting data pertaining to the relationship - family

### Removing duplicated values and noise caused by associations in the data

```python
import csv
```

```python
with open('relations.csv') as csvfile:
    reader = list(csv.reader(csvfile))
    family = []
    family_list = []
    family_set = set()
    for l in reader:
        if l[1] == 'Family':
            family.append(l)
    for f in family:
        if not(f[2]+f[0]) in family_set and not(f[0]+f[2]) in family_set:
            family_set.add(f[0]+f[2]) 
            family_list.append(f)
```

```python
print(len(family_list))
```

```python
print(family_list)
```

### Writing cleaned data into another csv

```python
with open("family.csv",'w') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerows(family_list)
```
